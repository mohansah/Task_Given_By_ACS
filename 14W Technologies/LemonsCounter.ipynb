```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt

# Function to get lemons count and segmented lemons
def LemonsCounter(file):
    # Reading the colour image
    color_img = cv2.imread(file)

    # Converting colour image to gray scale
    gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)

    # Performing binary thresholding on gray scale image
    ret, thresh = cv2.threshold(gray_img, 0 , 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Using a 5x5 kernel with full of ones for dilation operation
    kernel = np.ones((5,5),np.uint8)

    # Closing small holes inside the foreground objects, or small black points on the object
    closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    
    conts, h = cv2.findContours(closing, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Loop to count the number of lemons whose area is greater than one
    count = 0
    for cont in conts:
        area = cv2.contourArea(cont)
        if area>0:
            count+=1
            
    # create a mask with white pixels
    mask = np.ones(color_img.shape, dtype=np.uint8)
    mask.fill(255) # Replace 1 with 255
    
    # fill the contours point into the mask
    cv2.fillPoly(mask, conts, 0)

    # applying th mask to original image
    masked_image = cv2.bitwise_and(color_img, mask)
    
    # OpenCV read and display image with channel order B, G, R and Matplotlib read and display image with channel order R, G, B and
    # here we are using Matplotlib for displaying the image so we have to change the order from BGR to RGB
    FinalImage = cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB)
    
    return count, FinalImage
```

```python
StartTime = int(round(time.time() * 1000))
Count, SegmentedImage = LemonsCounter("lemon1.jpg")
EndTime = int(round(time.time() * 1000))

# Results
print("Number of Lemons:",Count)
print("Time Taken:", EndTime - StartTime,"milliseconds")
plt.title("Segmented Lemons")
plt.imshow(SegmentedImage)
plt.show()
```
